### Time Series plot with range slider
Plotly is one of a number of graphics packages that's available  
for Python, R and javascript. This example uses Python.  
  
The dataset contains over 10 years of Amazon stock prices.  
This example is a typical time series graph with the addition  
of a range slider on the x-axis.

DJIA 30 Stock Time Series  
https://www.kaggle.com/szrlee/stock-time-series-20050101-to-20171231

```python
import plotly
import pandas as pd
```

#### Offline graphs
There are a number of ways to create graphs in Plotly.  
This example displays offline and in a Jupyter notebook  
so the line of code below is needed to initialize notebook mode.

For more information:  
https://plot.ly/python/getting-started/

```python
plotly.offline.init_notebook_mode(connected=True)
```

#### Read Data  
The data is read from a csv file with the first  
five rows displayed below. Only the "Date" and  
"Close" fields are used in the graph.

```python
data = pd.read_csv('H:\\My Documents\\Work\\datasets\\AMZN.csv')
```

```python
data.head()
```

#### Graph  
The graph show Amazon stock prices from 2006 - 2017.  
I added a range slider to the x-axis and buttons that  
pre-select monthly intervals.  

The range selector makes it very easy to isolate  
specific time frames for closer inspection.

```python
#range yaxis defaults to the data --won't start at 0 by default

figure = {
    
    'data': [
        {
        'type': 'scatter',
        'x': data['Date'],
        'y': data['Close'],
        'opacity': 1.0,
        'line': {'color': 'rgb(60, 100, 160'}
        }
    ],
    
    'layout': {'title': 'AMZN Stock Price',
              'xaxis': {'rangeselector': {
                  'buttons': [{'count': 6, 'label': '6m', 'step': 'month', 'stepmode': 'backward'},
                             {'count': 12, 'label': '12m', 'step': 'month', 'stepmode': 'backward'},
                             {'count': 18, 'label': '18m', 'step': 'month', 'stepmode': 'backward'},
                             {'step': 'all'}]
              },
                                    
              'rangeslider': {}}, 
              'yaxis': {'title':'Stock Price', 'range': [0, 1300]}
               
              },
        
}
```

```python
plotly.offline.iplot(figure)
```
