### Heatmap with Plotly
Plotly is one of a number of graphics packages that's available  
for Python, R and javascript. This example uses Python and  
displays a heatmap.  
  
The dataset is from Kaggle.com  
Formula 1 Race Data  
https://www.kaggle.com/cjgdev/formula-1-race-data-19502017

```python
import plotly
import pandas as pd
```

#### Offline graphs
There are a number of ways to create graphs in Plotly.  
This example displays offline and in a Jupyter notebook  
so the line of code below is needed to initialize notebook mode.

For more information:  
https://plot.ly/python/getting-started/

```python
plotly.offline.init_notebook_mode(connected=True)
```

```python
datax = pd.read_csv('D:\\datasets\\graph_data\\hmap_test.csv', encoding='latin-1')
```

#### The Graph  
There are a few was to write the code for a graph.  
More details can be found on the Plotly site. Because graphs are specified  
in dictionary format in Python, I wrote this explicitly as a dictionary.

Plotly graphs have two sections: data and layout. The data section contains the  
graph type and data used while the layout section specifies much of the look  
of the graph. There's a good deal of customization that can be applied for each  
type of graph that API pages of the Plotly site addresses pretty well.

The colorscale in the heatmap uses the minimum and maximum rgb values instead  
of using a pre-defined colorscale. In general, very few parameters are needed  
to create quality graphs with Plotly.

```python
figure = {
    
    'data': [
        {
        'type': 'heatmap',
        'x': datax['race_name'],
        'y': datax['driver'],
        'z': datax['final_pos'],
        'colorscale': [[0, 'rgb(190,225,255)'], [1, 'rgb(35,70,110)']]
        }
    ],
    
    'layout': {'title': 'Driver Finish by Race',
              'xaxis': {'title': 'Race'},
               'yaxis': {'title': 'Driver'},
              'showlegend': False,
              'margin': {'l': 150, 'r': 75, 'b': 150},
              'width': 900,
              'height': 700
              }  
    
}
```

#### Ploting the Graph
The line of code below is used to plot the graph offline.  

All of the default controls are still present and a mouseover shows the data deails  
of the specific square. Light to dark values display the race finish order with light  
values starting at 1st place to the last place finisher at the max color of the colorscale.

From the graph it's easy to see that drivers like Vettel, Hamilton or Alonso had  
a pretty good season with finishes at the top end in most races while de la Rosa, Petrov  
or Karthikeyan finished near the bottom. The empty squares show a driver that didn't place  
in the points for a race.

Github doesn't display plotly graphics in a notebook so the graphic is displayed in the file: __F1_Heatmap.PNG__

```python
plotly.offline.iplot(figure)
```
